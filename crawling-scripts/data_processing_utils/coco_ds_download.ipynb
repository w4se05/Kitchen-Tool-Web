```python
### Download images utensils
import fiftyone as fo
import fiftyone.zoo as foz

dataset = foz.load_zoo_dataset(
    "coco-2017",
    split="train",
    label_types=["detections"],
    classes=["knife", "bowl"], # ["person"]
    max_samples=1200,
)

# 1. Define where you want the YOLO files to go
output_dir = "D:/CNN projects/YOLO/coco_2_background"

# 2. Define the classes you want to export 
# (It's good practice to list them again to ensure the YOLO ID order is fixed)
target_classes = ["knife", "bowl", "spoon", "fork"]

# 3. EXPORT directly to YOLOv5 / YOLOv8 format
print(f"Exporting to {output_dir}...")

dataset.export(
    export_dir=output_dir,
    dataset_type=fo.types.YOLOv5Dataset,  # This works for YOLOv8 too
    label_field="ground_truth",           # The standard field name for COCO zoo loads
    split="train",                        # Organizes into train/val folders
    classes=None,                         # Ensures only these classes are exported
)

print("âœ… Done! You can now point your YOLO model to the 'dataset.yaml' in that folder.")
```
